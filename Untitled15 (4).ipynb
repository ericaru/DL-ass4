```python
# Cell 1: Import necessary libraries
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.io import arff
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)
```

```python
# Cell 2: Load and explore the dataset
def load_arff_data(file_path):
    """Load ARFF file and convert to pandas DataFrame"""
    data, meta = arff.loadarff(file_path)
    df = pd.DataFrame(data)

    # Convert byte strings to regular strings
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].astype(str)

    return df, meta

# Load the dataset (assuming the file is named 'adult.arff')
df, meta = load_arff_data('adult.arff')

print("Dataset shape:", df.shape)
print("\nColumn names:")
print(df.columns.tolist())
print("\nFirst few rows:")
print(df.head())
print("\nData types:")
print(df.dtypes)
print("\nTarget variable distribution:")
print(df['income'].value_counts())
```

```python
# Cell 3: Data preprocessing function with visualization
def preprocess_data(df):
    """
    Preprocess the Adult dataset:
    - Handle missing values
    - Group rare categorical values to help GAN training
    - Encode categorical variables
    - Normalize continuous features
    - Split into train/test sets
    """

    # Make a copy to avoid modifying original data
    data = df.copy()

    # Replace '?' with NaN for missing values
    data = data.replace('?', np.nan)

    # Handle missing values by dropping rows with missing values
    # (Alternative: could use mode/median imputation)
    data = data.dropna()

    # Store original data for comparison plots
    data_original = data.copy()

    # Group rare categorical values to help GAN training
    def group_rare_categories(series, min_frequency=0.02):
        """Group categories that appear less than min_frequency of the time into 'Other'"""
        value_counts = series.value_counts(normalize=True)
        rare_categories = value_counts[value_counts < min_frequency].index
        series_grouped = series.copy()
        series_grouped[series_grouped.isin(rare_categories)] = 'Other'
        return series_grouped

    # Apply rare category grouping to categorical columns
    categorical_to_group = ['workclass', 'occupation', 'native-country']
    preprocessing_changes = {}

    for col in categorical_to_group:
        if col in data.columns:
            original_unique = data[col].nunique()
            preprocessing_changes[col] = {
                'original': data_original[col].value_counts(),
                'processed': None
            }
            data[col] = group_rare_categories(data[col])
            preprocessing_changes[col]['processed'] = data[col].value_counts()
            new_unique = data[col].nunique()
            print(f"{col}: {original_unique} -> {new_unique} categories")

    # Handle capital-gain and capital-loss by binning (they are heavily skewed)
    if 'capital-gain' in data.columns:
        preprocessing_changes['capital-gain'] = {
            'original': data_original['capital-gain'].value_counts().head(10),
            'processed': None
        }
        # Most values are 0, bin the rest
        data['capital-gain-binned'] = pd.cut(data['capital-gain'],
                                           bins=[-1, 0, 1000, 5000, float('inf')],
                                           labels=['zero', 'low', 'medium', 'high'])
        preprocessing_changes['capital-gain']['processed'] = data['capital-gain-binned'].value_counts()
        data = data.drop('capital-gain', axis=1)
        print(f"capital-gain binned into: {data['capital-gain-binned'].value_counts()}")

    if 'capital-loss' in data.columns:
        preprocessing_changes['capital-loss'] = {
            'original': data_original['capital-loss'].value_counts().head(10),
            'processed': None
        }
        # Most values are 0, bin the rest
        data['capital-loss-binned'] = pd.cut(data['capital-loss'],
                                           bins=[-1, 0, 1000, 2000, float('inf')],
                                           labels=['zero', 'low', 'medium', 'high'])
        preprocessing_changes['capital-loss']['processed'] = data['capital-loss-binned'].value_counts()
        data = data.drop('capital-loss', axis=1)
        print(f"capital-loss binned into: {data['capital-loss-binned'].value_counts()}")

    # Plot before and after comparison
    plot_preprocessing_changes(preprocessing_changes)

    # Identify categorical and numerical columns after preprocessing
    categorical_columns = []
    numerical_columns = []

    for col in data.columns:
        if col == 'income':  # Skip target variable
            continue
        if data[col].dtype == 'object' or col.endswith('-binned'):
            categorical_columns.append(col)
        else:
            numerical_columns.append(col)

    print(f"Categorical columns: {categorical_columns}")
    print(f"Numerical columns: {numerical_columns}")

    # Encode target variable
    le_target = LabelEncoder()
    data['income_encoded'] = le_target.fit_transform(data['income'])

    # One-hot encode categorical variables
    data_encoded = pd.get_dummies(data, columns=categorical_columns, prefix=categorical_columns)

    # Separate features and target
    X = data_encoded.drop(['income', 'income_encoded'], axis=1)
    y = data_encoded['income_encoded']

    # Normalize numerical features
    scaler = StandardScaler()
    if numerical_columns:
        X[numerical_columns] = scaler.fit_transform(X[numerical_columns])

    return X, y, scaler, le_target, categorical_columns, numerical_columns

def plot_preprocessing_changes(changes_dict):
    """Plot before and after preprocessing changes"""
    n_features = len(changes_dict)
    fig, axes = plt.subplots(n_features, 2, figsize=(15, 4*n_features))

    if n_features == 1:
        axes = axes.reshape(1, -1)

    for i, (feature, data) in enumerate(changes_dict.items()):
        # Original distribution
        original_counts = data['original']
        processed_counts = data['processed']

        # Plot original
        ax_orig = axes[i, 0]
        original_counts.plot(kind='bar', ax=ax_orig, color='lightcoral', alpha=0.7)
        ax_orig.set_title(f'{feature} - Original Distribution\n({len(original_counts)} categories)')
        ax_orig.set_xlabel('Categories')
        ax_orig.set_ylabel('Count')
        ax_orig.tick_params(axis='x', rotation=45)

        # Plot processed
        ax_proc = axes[i, 1]
        processed_counts.plot(kind='bar', ax=ax_proc, color='lightblue', alpha=0.7)
        ax_proc.set_title(f'{feature} - After Preprocessing\n({len(processed_counts)} categories)')
        ax_proc.set_xlabel('Categories')
        ax_proc.set_ylabel('Count')
        ax_proc.tick_params(axis='x', rotation=45)

        # Add count labels on bars for processed data (since it's fewer categories)
        for j, v in enumerate(processed_counts.values):
            ax_proc.text(j, v + max(processed_counts.values)*0.01, str(v),
                        ha='center', va='bottom', fontsize=9)

    plt.tight_layout()
    plt.show()

    # Print summary of changes
    print("\n" + "="*60)
    print("PREPROCESSING SUMMARY:")
    print("="*60)
    for feature, data in changes_dict.items():
        original_cats = len(data['original'])
        processed_cats = len(data['processed'])
        reduction = original_cats - processed_cats
        print(f"{feature}:")
        print(f"  Categories: {original_cats} → {processed_cats} (reduced by {reduction})")
        if 'Other' in data['processed'].index:
            other_count = data['processed']['Other']
            total_count = data['processed'].sum()
            other_pct = (other_count / total_count) * 100
            print(f"  'Other' category: {other_count} samples ({other_pct:.1f}%)")
        print()
```

```python
# Cell 4: Apply preprocessing and create train/test splits
def create_train_test_splits(X, y, test_size=0.2, random_state=42):
    """Create stratified train/test split maintaining label ratios"""
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state, stratify=y
    )

    return X_train, X_test, y_train, y_test

# Process the data
X, y, scaler, le_target, categorical_columns, numerical_columns = preprocess_data(df)

print(f"\nProcessed data shape: {X.shape}")
print(f"Feature columns: {X.columns.tolist()}")
print(f"Target distribution: {y.value_counts()}")
```

# GAN

```python
# Cell 1: Define Autoencoder for data embedding (as specified in assignment)
class Autoencoder(nn.Module):
    def __init__(self, input_dim, embedding_dim=32):
        super(Autoencoder, self).__init__()

        # Encoder
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, input_dim // 2),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(input_dim // 2, embedding_dim),
            nn.ReLU()
        )

        # Decoder
        self.decoder = nn.Sequential(
            nn.Linear(embedding_dim, input_dim // 2),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(input_dim // 2, input_dim),
            nn.Sigmoid()  # To match normalized input range
        )

    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded, encoded

    def encode(self, x):
        return self.encoder(x)
```

```python
# Cell 2: Define GAN components
class Generator(nn.Module):
    def __init__(self, noise_dim, embedding_dim, hidden_dims=[128, 256]):
        super(Generator, self).__init__()

        layers = []
        input_dim = noise_dim

        # Hidden layers
        for hidden_dim in hidden_dims:
            layers.extend([
                nn.Linear(input_dim, hidden_dim),
                nn.BatchNorm1d(hidden_dim),
                nn.ReLU(),
                nn.Dropout(0.3)
            ])
            input_dim = hidden_dim

        # Output layer
        layers.extend([
            nn.Linear(input_dim, embedding_dim),
            nn.Tanh()  # To match autoencoder embedding range
        ])

        self.model = nn.Sequential(*layers)

    def forward(self, noise):
        return self.model(noise)

class Discriminator(nn.Module):
    def __init__(self, embedding_dim, hidden_dims=[256, 128]):
        super(Discriminator, self).__init__()

        layers = []
        input_dim = embedding_dim

        # Hidden layers
        for hidden_dim in hidden_dims:
            layers.extend([
                nn.Linear(input_dim, hidden_dim),
                nn.LeakyReLU(0.2),
                nn.Dropout(0.3)
            ])
            input_dim = hidden_dim

        # Output layer
        layers.extend([
            nn.Linear(input_dim, 1),
            nn.Sigmoid()
        ])

        self.model = nn.Sequential(*layers)

    def forward(self, x):
        return self.model(x)
```

Training functions


```python
# Cell 3: Training functions
def train_autoencoder(autoencoder, data_loader, num_epochs=350, lr=0.001, device='cpu'):
    """Train the autoencoder with early stopping"""
    print(f"DEBUG: Training autoencoder for {num_epochs} epochs")  # Add this line
    criterion = nn.MSELoss()
    optimizer = optim.Adam(autoencoder.parameters(), lr=lr)

    autoencoder.train()
    losses = []
    best_loss = float('inf')
    patience = 20  # Increased patience for more thorough training
    patience_counter = 0

    print("Training Autoencoder...")
    for epoch in range(num_epochs):
        epoch_loss = 0
        for batch_data in data_loader:
            data = batch_data[0].to(device)  # Only features, not labels

            optimizer.zero_grad()
            reconstructed, _ = autoencoder(data)
            loss = criterion(reconstructed, data)
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()

        avg_loss = epoch_loss / len(data_loader)
        losses.append(avg_loss)

        # Early stopping with improved criteria
        if avg_loss < best_loss - 1e-6:  # Require minimum improvement
            best_loss = avg_loss
            patience_counter = 0
        else:
            patience_counter += 1

        if (epoch + 1) % 25 == 0:  # Report every 25 epochs
            print(f'  AE Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}, Best: {best_loss:.4f}')

        if patience_counter >= patience:
            print(f'  AE Early stopping at epoch {epoch+1}, best loss: {best_loss:.4f}')
            break

    if patience_counter < patience:
        print(f'  AE Training completed at epoch {num_epochs}, final loss: {avg_loss:.4f}')

    return losses
```

```python

# CELL 3: GAN Training - Update Discriminator Every 5 Batches
def train_gan_batch_based_updates(generator, discriminator, autoencoder, data_loader, config, device='cpu'):
    """Train generator every batch, discriminator every N batches"""

    opt_G = optim.Adam(generator.parameters(), lr=config['lr_g'], betas=(0.5, 0.999))
    opt_D = optim.Adam(discriminator.parameters(), lr=config['lr_d'], betas=(0.5, 0.999))

    G_losses = []
    D_losses = []

    # Discriminator update frequency (every N batches)
    discriminator_batch_frequency = config.get('d_batch_frequency', 5)  # Update D every 5 batches

    print(f"  Generator: every batch, Discriminator: every {discriminator_batch_frequency} batches")

    # Generator head start training (if specified)
    if config.get('generator_head_start_epochs', 0) > 0:
        print(f"  Generator head start: {config['generator_head_start_epochs']} epochs")
        for head_start_epoch in range(config['generator_head_start_epochs']):
            epoch_G_loss = 0

            for batch_data in data_loader:
                batch_size = batch_data[0].size(0)
                noise = torch.randn(batch_size, config['noise_dim']).to(device)
                fake_embeddings = generator(noise)

                with torch.no_grad():
                    real_data = batch_data[0].to(device)
                    _, real_embeddings = autoencoder(real_data)
                    real_mean = real_embeddings.mean(dim=0, keepdim=True)
                    real_std = real_embeddings.std(dim=0, keepdim=True)
                    fake_normalized = (fake_embeddings - fake_embeddings.mean(dim=0, keepdim=True)) / (fake_embeddings.std(dim=0, keepdim=True) + 1e-8)
                    target_embeddings = fake_normalized * real_std + real_mean

                opt_G.zero_grad()
                g_loss = nn.MSELoss()(fake_embeddings, target_embeddings)
                g_loss.backward()
                opt_G.step()

                epoch_G_loss += g_loss.item()

            if (head_start_epoch + 1) % 5 == 0:
                avg_g_loss = epoch_G_loss / len(data_loader)
                print(f"    Head start epoch {head_start_epoch + 1}: G_loss = {avg_g_loss:.4f}")

    # Main adversarial training loop
    print(f"  Starting adversarial training...")

    global_batch_count = 0  # Track total batches across all epochs

    for epoch in range(config['num_epochs']):
        epoch_G_loss = 0
        epoch_D_loss = 0
        d_updates_this_epoch = 0

        for step, batch_data in enumerate(data_loader):
            batch_size = batch_data[0].size(0)
            real_data = batch_data[0].to(device)

            # Get real embeddings from autoencoder
            with torch.no_grad():
                _, real_embeddings = autoencoder(real_data)

            # Generate fake embeddings
            noise = torch.randn(batch_size, config['noise_dim']).to(device)
            fake_embeddings = generator(noise)

            # ========================================
            # DISCRIMINATOR UPDATE (every N batches)
            # ========================================

            update_discriminator = (global_batch_count % discriminator_batch_frequency == 0)

            if update_discriminator:
                opt_D.zero_grad()

                # Train discriminator on REAL embeddings
                disc_real_output = discriminator(real_embeddings)
                real_labels = torch.ones(batch_size, 1).to(device)

                # Apply label smoothing if specified
                if config.get('real_label_smooth', 1.0) < 1.0:
                    real_labels = real_labels * config['real_label_smooth']

                disc_loss_real = nn.BCELoss()(disc_real_output, real_labels)

                # Train discriminator on FAKE embeddings (with detach!)
                disc_fake_output = discriminator(fake_embeddings.detach())
                fake_labels = torch.zeros(batch_size, 1).to(device)

                if config.get('fake_label_smooth', 0.0) > 0.0:
                    fake_labels = fake_labels + config['fake_label_smooth']

                disc_loss_fake = nn.BCELoss()(disc_fake_output, fake_labels)

                # Total discriminator loss
                disc_loss = disc_loss_real + disc_loss_fake
                disc_loss.backward()

                # Optional gradient clipping
                if config.get('d_grad_clip', 0) > 0:
                    torch.nn.utils.clip_grad_norm_(discriminator.parameters(), config['d_grad_clip'])

                opt_D.step()
                d_updates_this_epoch += 1

                epoch_D_loss += disc_loss.item()

            else:
                # Still calculate discriminator loss for monitoring, but don't update
                with torch.no_grad():
                    disc_real_output = discriminator(real_embeddings)
                    disc_fake_output = discriminator(fake_embeddings.detach())
                    real_labels = torch.ones(batch_size, 1).to(device)
                    fake_labels = torch.zeros(batch_size, 1).to(device)

                    if config.get('real_label_smooth', 1.0) < 1.0:
                        real_labels = real_labels * config['real_label_smooth']
                    if config.get('fake_label_smooth', 0.0) > 0.0:
                        fake_labels = fake_labels + config['fake_label_smooth']

                    disc_loss = (nn.BCELoss()(disc_real_output, real_labels) +
                               nn.BCELoss()(disc_fake_output, fake_labels))
                    epoch_D_loss += disc_loss.item()

            # ========================================
            # GENERATOR UPDATE (every batch)
            # ========================================

            opt_G.zero_grad()

            # Generate NEW fake embeddings for generator training
            noise = torch.randn(batch_size, config['noise_dim']).to(device)
            fake_embeddings = generator(noise)

            # Goal: fool discriminator into outputting high probability
            disc_fake_output = discriminator(fake_embeddings)
            target_labels = torch.ones(batch_size, 1).to(device)

            gen_loss = nn.BCELoss()(disc_fake_output, target_labels)
            gen_loss.backward()

            # Optional gradient clipping
            if config.get('g_grad_clip', 0) > 0:
                torch.nn.utils.clip_grad_norm_(generator.parameters(), config['g_grad_clip'])

            opt_G.step()

            epoch_G_loss += gen_loss.item()
            global_batch_count += 1

            # Print progress every 50 steps, showing update status
            if step % 50 == 0:
                d_status = "UPDATED" if update_discriminator else "skipped"
                print(f'    Epoch: {epoch} Step: {step} G_Loss: {gen_loss.item():.4f} '
                      f'D_Loss: {disc_loss.item():.4f} (D: {d_status})')

        # Average losses for the epoch
        avg_G_loss = epoch_G_loss / len(data_loader)
        # For D_loss, average only over the batches where D was updated
        avg_D_loss = epoch_D_loss / max(1, d_updates_this_epoch + (len(data_loader) - d_updates_this_epoch))

        G_losses.append(avg_G_loss)
        D_losses.append(avg_D_loss)

        # Simple early stopping
        should_stop, stop_reason = check_simple_early_stopping_batch_based(
            avg_G_loss, avg_D_loss, epoch, config, d_updates_this_epoch, len(data_loader)
        )

        if (epoch + 1) % 10 == 0:
            print(f'    Epoch [{epoch+1}/{config["num_epochs"]}] G_Loss: {avg_G_loss:.4f}, '
                  f'D_Loss: {avg_D_loss:.4f} (D updated: {d_updates_this_epoch}/{len(data_loader)} batches)')

        if should_stop:
            print(f'    Early stopping at epoch {epoch+1}: {stop_reason}')
            break

    # Return best model state (last epoch)
    best_model_state = {
        'generator': generator.state_dict(),
        'discriminator': discriminator.state_dict(),
        'epoch': len(G_losses) - 1,
        'g_loss': G_losses[-1] if G_losses else float('inf'),
        'd_loss': D_losses[-1] if D_losses else float('inf')
    }

    return G_losses, D_losses, best_model_state

def check_simple_early_stopping_batch_based(g_loss, d_loss, epoch, config, d_updates, total_batches):
    """Simple early stopping adapted for batch-based discriminator updates"""

    # 1. Mode collapse: G_loss very low
    if g_loss < 0.05:
        return True, "Potential mode collapse (G_loss too low)"

    # 2. Generator dominance: G_loss very low and D not learning
    if g_loss < 0.1 and d_loss > 3.0:
        return True, "Generator dominance"

    # 3. Training instability: both losses very high
    if g_loss > 10.0 and d_loss > 10.0:
        return True, "Training divergence"

    # 4. Generator can't learn: G_loss stuck very high
    if epoch > 30 and g_loss > 5.0:
        return True, "Generator unable to learn"

    # 5. Maximum epochs
    if epoch >= config['num_epochs'] - 1:
        return True, "Maximum epochs reached"

    return False, ""
```

```python
# Cell 5: Updated hyperparameter search for batch-based discriminator updates
def find_best_hyperparameters_batch_based(X_train_tensor, y_train_tensor, preprocessing_info, test_seed=42):
    """Find best hyperparameters using batch-based discriminator updates"""

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")
    print(f"Testing hyperparameters on seed {test_seed}")

    # Set seed for hyperparameter search
    torch.manual_seed(test_seed)
    np.random.seed(test_seed)

    # Use the new batch-based configurations instead of the old ones
    configs = get_batch_based_configs()  # Changed from get_hyperparameter_configs()
    results = []

    print(f"\nTesting {len(configs)} batch-based configurations...")

    for i, config in enumerate(configs):
        print(f"\n[{i+1}/{len(configs)}] Testing {config['name']}")
        print(f"  LR_G: {config['lr_g']}, LR_D: {config['lr_d']}")
        print(f"  Embedding: {config['embedding_dim']}, Noise: {config['noise_dim']}")
        print(f"  Batch: {config['batch_size']}, D_freq: {config['d_batch_frequency']}")  # Added D frequency
        print(f"  G_head_start: {config['generator_head_start_epochs']}, Patience: {config['patience']}")

        try:
            # Create data loader
            dataset = TensorDataset(X_train_tensor, y_train_tensor)
            data_loader = DataLoader(dataset, batch_size=config['batch_size'], shuffle=True)

            # Initialize models
            input_dim = preprocessing_info['feature_dim']

            autoencoder = Autoencoder(input_dim, config['embedding_dim']).to(device)
            generator = Generator(config['noise_dim'], config['embedding_dim'],
                                config['gen_hidden']).to(device)
            discriminator = Discriminator(config['embedding_dim'],
                                        config['disc_hidden']).to(device)

            # Train autoencoder with more epochs and early stopping
            ae_losses = train_autoencoder(autoencoder, data_loader, num_epochs=350, device=device)
            ae_final_loss = ae_losses[-1]

            # Train GAN using the NEW batch-based training function
            G_losses, D_losses, best_model_state = train_gan_batch_based_updates(
                generator, discriminator, autoencoder, data_loader, config, device
            )

            # Calculate evaluation metrics for this configuration
            min_g_loss = min(G_losses) if G_losses else float('inf')
            final_g_loss = G_losses[-1] if G_losses else float('inf')
            avg_g_loss_last_10 = np.mean(G_losses[-10:]) if len(G_losses) >= 10 else final_g_loss

            # Stability metric: standard deviation of last 20% of training
            last_20_percent = max(1, len(G_losses) // 5)
            stability = np.std(G_losses[-last_20_percent:]) if len(G_losses) > 5 else float('inf')

            # Convergence speed: how quickly it reached 90% of best loss
            convergence_epoch = len(G_losses)
            target_loss = min_g_loss * 1.1  # 110% of best loss
            for epoch, loss in enumerate(G_losses):
                if loss <= target_loss:
                    convergence_epoch = epoch
                    break

            # Combined score (lower is better)
            # Weights: 40% final performance, 30% stability, 20% convergence speed, 10% autoencoder
            combined_score = (0.4 * min_g_loss +
                            0.3 * stability +
                            0.2 * (convergence_epoch / len(G_losses)) +
                            0.1 * ae_final_loss)

            result = {
                'config': config,
                'min_g_loss': min_g_loss,
                'final_g_loss': final_g_loss,
                'avg_g_loss_last_10': avg_g_loss_last_10,
                'stability': stability,
                'convergence_epoch': convergence_epoch,
                'convergence_ratio': convergence_epoch / len(G_losses),
                'ae_final_loss': ae_final_loss,
                'combined_score': combined_score,
                'total_epochs': len(G_losses),
                'early_stopped': len(G_losses) < config['num_epochs'],
                'd_batch_frequency': config['d_batch_frequency']  # Track this important parameter
            }

            results.append(result)

            print(f"  Results: Min G_loss: {min_g_loss:.4f}, Stability: {stability:.4f}")
            print(f"  Combined Score: {combined_score:.4f}, Epochs: {len(G_losses)}")

        except Exception as e:
            print(f"  ERROR in {config['name']}: {str(e)}")
            continue

    # Find best configuration based on combined score
    if not results:
        raise ValueError("No configurations completed successfully!")

    best_result = min(results, key=lambda x: x['combined_score'])

    print(f"\n{'='*60}")
    print("BATCH-BASED HYPERPARAMETER SEARCH RESULTS")
    print(f"{'='*60}")
    print(f"Best Configuration: {best_result['config']['name']}")
    print(f"Combined Score: {best_result['combined_score']:.4f}")
    print(f"Min Generator Loss: {best_result['min_g_loss']:.4f}")
    print(f"Training Stability (std): {best_result['stability']:.4f}")
    print(f"Convergence Speed: {best_result['convergence_epoch']}/{best_result['total_epochs']} epochs")
    print(f"Discriminator Update Frequency: every {best_result['d_batch_frequency']} batches")  # New info
    print(f"Early Stopped: {best_result['early_stopped']}")
    print(f"\nBest Config Parameters:")
    for key, value in best_result['config'].items():
        if key != 'name':
            print(f"  {key}: {value}")

    # Show top 5 configurations
    top_5 = sorted(results, key=lambda x: x['combined_score'])[:5]
    print(f"\nTop 5 Configurations:")
    for i, result in enumerate(top_5):
        print(f"{i+1}. {result['config']['name']}: Score={result['combined_score']:.4f}, "
              f"Min_G_Loss={result['min_g_loss']:.4f}, D_freq={result['d_batch_frequency']}")

    return best_result['config'], results

def train_final_models_with_best_config_batch_based(data_splits, preprocessing_info, best_config, seeds=[42, 7, 17]):
    """Train final models using the best batch-based configuration for each seed"""

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"\n{'='*60}")
    print("TRAINING FINAL MODELS WITH BEST BATCH-BASED CONFIGURATION")
    print(f"{'='*60}")
    print(f"Using configuration: {best_config['name']}")
    print(f"Discriminator update frequency: every {best_config['d_batch_frequency']} batches")

    final_results = {}

    for seed_idx, seed in enumerate(seeds):
        print(f"\nTraining final model for seed {seed}...")

        # Set seed
        torch.manual_seed(seed)
        np.random.seed(seed)

        # Get data for this seed
        X_train, X_test, y_train, y_test = data_splits[seed_idx]
        X_train_tensor, X_test_tensor, y_train_tensor, y_test_tensor = prepare_pytorch_data(
            X_train, X_test, y_train, y_test
        )

        # Create data loader
        dataset = TensorDataset(X_train_tensor, y_train_tensor)
        data_loader = DataLoader(dataset, batch_size=best_config['batch_size'], shuffle=True)

        # Initialize models
        input_dim = preprocessing_info['feature_dim']

        autoencoder = Autoencoder(input_dim, best_config['embedding_dim']).to(device)
        generator = Generator(best_config['noise_dim'], best_config['embedding_dim'],
                            best_config['gen_hidden']).to(device)
        discriminator = Discriminator(best_config['embedding_dim'],
                                    best_config['disc_hidden']).to(device)

        # Train autoencoder with extended training
        print("  Training autoencoder...")
        ae_losses = train_autoencoder(autoencoder, data_loader, num_epochs=350, device=device)

        # Train GAN using batch-based updates
        print("  Training GAN with batch-based discriminator updates...")
        G_losses, D_losses, best_model_state = train_gan_batch_based_updates(
            generator, discriminator, autoencoder, data_loader, best_config, device
        )

        # Save model
        model_path = f'/content/best_batch_based_gan_model_seed_{seed}.pt'
        torch.save({
            'seed': seed,
            'config': best_config,
            'autoencoder_state': autoencoder.state_dict(),
            'generator_state': best_model_state['generator'],
            'discriminator_state': best_model_state['discriminator'],
            'preprocessing_info': preprocessing_info,
            'final_g_loss': min(G_losses) if G_losses else float('inf'),
            'training_history': {
                'g_losses': G_losses,
                'd_losses': D_losses,
                'ae_losses': ae_losses
            }
        }, model_path)

        final_results[seed] = {
            'config': best_config,
            'g_losses': G_losses,
            'd_losses': D_losses,
            'ae_losses': ae_losses,
            'best_model_state': best_model_state,
            'final_g_loss': min(G_losses) if G_losses else float('inf'),
            'model_path': model_path
        }

        print(f"  Model saved to: {model_path}")
        print(f"  Final G_loss: {final_results[seed]['final_g_loss']:.4f}")
        print(f"  Training epochs: {len(G_losses)}")

    return final_results
```
